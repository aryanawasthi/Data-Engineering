```scala
//**Spark Streaming program to find the sum of each number entered from the socket

// importing all the necessary Imports
import org.apache.spark.SparkContext 
import org.apache.spark._ 
import org.apache.spark.streaming._ 
import org.apache.spark.streaming.StreamingContext._ 
// creating a spark context
val sc=new SparkContext("local[2]","wordcount")
// creating a streaming context
val ssc=new StreamingContext(sc,Seconds(2)) 
// creating a checkpoint directory
ssc.checkpoint(".")
// connecting to a local port as producer
val lines=ssc.socketTextStream("localhost",8988)
// using the reduceByWindow functions
// defining the summary function
def summary_func(x:String,y:String):String={
    (x.toInt+y.toInt).toString
}
// defining the inverse function
def inverse_func(x:String,y:String)={
    (x.toInt-y.toInt).toString
}
// defining count by using reduceByWindow function 
val count=lines.reduceByWindow(summary_func,inverse_func,Seconds(10),Seconds(4))
// collecting the results
count.print()
// starting the streaming service
ssc.start()
ssc.awaitTermination()
```

```scala
//**Spark Streaming program to find the count the number of lines for  each number entered from the socket

// importing all the necessary Imports
import org.apache.spark.SparkContext 
import org.apache.spark._ 
import org.apache.spark.streaming._ 
import org.apache.spark.streaming.StreamingContext._ 
// creating a spark context
val sc=new SparkContext("local[2]","wordcount")
// creating a streaming context
val ssc=new StreamingContext(sc,Seconds(2)) 
// creating a checkpoint directory
ssc.checkpoint(".")
// connecting to a local port as producer
val lines=ssc.socketTextStream("localhost",8988)
// using the reduceByWindow functions

// defining count by using reduceByWindow function 
val count=lines.countByWindow(Seconds(10),Seconds(4))
// collecting the results
count.print()
// starting the streaming service
ssc.start()
ssc.awaitTermination()
```
